## Simple implementation of grid search on finding the optima for logistic regression

```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv('data.csv', header = None)
```

```python
df   # take a look at the dataframe; we only need the 0th and 2nd column.
```

## Now do a grid search on the minimum combination 

Here is pic for it 


```python
leftSum = 0
rightSum = 0
totalSum = 0
optimum = 10000000  # we are finding the minimum one; so set the default optimum value to be some high num

b0 = -9
b1 = 0 
res = (0, 0)
while b0 <= -8:
    while b1 <= 1:
        leftSum = 0
        rightSum = 0
        totalSum = 0
        for i, v in enumerate(df[0]):
            leftSum = (b0 + b1 * v) * df[2][i]
            rightSum = math.log(1 + math.exp(b0 + b1 * v))
            totalSum += (leftSum - rightSum)
        totalSum = np.multiply(totalSum, -2)
        if optimum > totalSum:
            optimum = min(totalSum, optimum)
            res = (b0, b1)
        b1 += 0.1
    b0 += 0.1
    b1 = 0

    
print ('The minimum value is {}'.format(optimum))
print ('And the optimal b0 and b1 values are {} and {}, respectively.'.format(res[0], res[1]))
```
